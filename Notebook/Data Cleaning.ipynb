# Data Cleaning.ipynb

### Jupyter Notebook created to do the Part 2 of the project. This .ipynb file will do the data cleaning and handle missing/null values.

## Importing the Required Libraries

#### Pandas : Pandas is the only library imported here because we only need to read the .csv file and do the data cleaning.

```python
import pandas as pd
```

## Importing the CSV File

```python
df = pd.read_csv("D:\Autonomous Driving Deep Learning Project\Datasets & Problem Statement\Part 2\Tesla - Deaths.csv")
df.head()
```

## Basic Info of the Dataset

```python
df.shape
df.info()
df.isnull().sum()
```

## Naming the Unnamed Columns

```python
df = df.rename(columns={"Unnamed: 16": "Link","Unnamed: 17": "News Link"})
df.head()
```

## Removing Fully Empty Rows

```python
df = df.dropna(how='all')
df.info
```

## Converting the Date Column into a single format

```python
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

## Saving the Clean CSV File

```python
df.to_csv("../data/tesla_cleaned.csv", index=False)
```
